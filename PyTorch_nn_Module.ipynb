```python
import torch
import torch.nn as nn
```

```python
class Model_1(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 1) #(input, output)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

```python
features = torch.rand(10,5)

model_1 = Model_1(features.shape[1])

model_1.forward(features)
```

```python
class Model_2(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear1 = nn.Linear(num_features, 3) #(input, output)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
```

```python
features = torch.rand(10,5)

model_2 = Model_2(features.shape[1])

model_2.forward(features)
```

```python
class Model_3(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3), #(input, output)
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):

    out = self.network(features)

    return out
```

```python
features = torch.rand(10,5)

model_3 = Model_3(features.shape[1])

model_3.forward(features)
```
